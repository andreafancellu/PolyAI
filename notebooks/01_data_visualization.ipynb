# PolyAI - Data Exploration and Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"..\data\polyai_generated_fake_dataset.csv")

print("Dataset shape:", df.shape)
print(df.head())
```

### Distribution of recycled percentage

```python
plt.figure(figsize=(8,5))
sns.histplot(df["recycled_percentage"], bins=20, color="steelblue")
plt.title("Distribution of Recycled Percentage")
plt.xlabel("Recycled Percentage (%)")
plt.ylabel("Count")
plt.show()
```

### Average recycled percentage by polymer type

```python
avg_recycled = df[df['recycled_percentage'] > 0].groupby("polymer_type")["recycled_percentage"].mean().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(x="polymer_type", y="recycled_percentage", data=avg_recycled, palette="viridis")
plt.title("Average Recycled Content by Polymer Type")
plt.xlabel("Polymer Type")
plt.ylabel("Average Recycled Percentage (%)")
plt.show()
```
